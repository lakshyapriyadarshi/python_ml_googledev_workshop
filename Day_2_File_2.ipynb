```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde= False)
```

```python
sns.distplot(tips['total_bill'],kde= False,bins=50)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='hot')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True,cmap='hot')
```

```python
sns.heatmap(tc,annot=True,cmap='cool')
```

```python
sns.heatmap(tc,robust=True,annot=True,cmap='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','x'])
```
